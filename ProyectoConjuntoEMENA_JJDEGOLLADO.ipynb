**Código de solución estudiante 1**

Primero se tiene que obtener c
$$c=\int_{-\infty}^{\infty}t(x)dx$$
El rango de nuestra función es de 0 a 1
$$c=\int_{0}^{1}t(x)dx$$
Con nuestra t(x) queda
$$c=\int_{0}^{1}{asin(\pi x)}dx$$
Sustituimos $u=\pi x$ y queda $du=\pi dx$ despejando el $dx$ queda $dx=\frac{du}{\pi}$. Ahora ponemos la integral en función de u y du.
$$c=\int_{0}^{1}{asin(u)}\frac{du}{\pi}$$
Sacamos las constantes
$$c=(\frac{a}{\pi})(\int_{0}^{1}{sin(u)}du)$$
Realizamos la integral
$$c=(\frac{a}{\pi})(-cos(u) \Big|_0^1)$$
Regresamos el valor de u
$$c=(\frac{a}{\pi})(-cos(\pi x) \Big|_0^1)$$
Evaluamos en 1 y 0
$$c=(\frac{a}{\pi})(-cos(\pi))-(\frac{a}{\pi})(-cos(0))$$
Nos queda c igual a
$$c=-\frac{acos(\pi)}{\pi} + \frac{acos(0)}{\pi}$$
Ahora para encontrar $g(x)$
$$g(x)=\frac{t(x)}{c}$$ 
Nuestra $t(x)$ es
$$t(x)=a\sin(\pi x)$$
Sustituimos $t(x) y c$
$$g(x)=\frac{a\sin(\pi x)}{-\frac{acos(\pi)}{\pi} + \frac{acos(0)}{\pi}}$$ 
Igualamos el denominador del denominador
$$g(x)=\frac{a\sin(\pi x)}{-\frac{acos(\pi)+acos(0)}{\pi}}$$ 
Pasamos $\pi$ al numerador
$$g(x)=\frac{\pi a\sin(\pi x)}{-acos(\pi)+acos(0)}$$ 
Factor común a en el denominador
$$g(x)=\frac{\pi a\sin(\pi x)}{a(-cos(\pi)+cos(0))}$$ 
Se eliminan las a
$$g(x)=\frac{\pi \sin(\pi x)}{-cos(\pi)+cos(0)}$$

### Ejercicio 6:
6. Comparar el porcentaje de puntos de aceptación cuando se usa $t(x)$ constante y $t(x)$ un pulso senoidal. Concluir

**Código de solución estudiante 1**

```python
# Función de aceptación rechazo
def prob_acep(R2:'Variables distruidas U~U(0,1)',
                 R1:'Variables distribuidas como g(x)',
                 f:'función objetivo a generar',
                 t:'función que mayora a f'):
    f_x = f(R1)
    t_x = t(R1)
    condition = R2*t_x <=f_x
    return sum(condition)
# Solución
# Función objetivo 
f = lambda x: 30 *(x**2 - 2*x**3 + x**4)
# Función t(x) senoidal
t1 = lambda x: a * np.sin(np.pi * x)
# Función t(x) constante
t2 = lambda x: a * np.ones(len(x))
# Simulaciones
N = 10000
R1 = np.random.rand(N)
R2 = np.random.rand(N)
# Probabilidad senoidal
p1 = prob_acep(R2, R1, f, t1)/N
# Probabilidad constante
p2 = prob_acep(R2, R1, f, t2)/N
print(f'La probabilidad de aceptación de puntos con t(x) senoidal es de {p1} y con t(x) constante es de {p2}')
```
